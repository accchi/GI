```python
import pysam
```

```python
total_number_of_reads = 0
zero_mapping_quality_counter = 0
mapping_quality_sum = 0
mapping_quality_sum_without_zero_quality = 0
unmapped_flag = 0x4
unmapped_reads_count= 0

with pysam.AlignmentFile('/sbgenomics/project-files/merged-tumor.bam') as alignment_file:
    for read in alignment_file:
        total_number_of_reads += 1
        mapping_quality_sum += read.mapping_quality
        
        if (read.mapping_quality == 0):
            zero_mapping_quality_counter += 1
            
        if (read.flag & unmapped_flag):
            unmapped_reads_count += 1
```

```python
print('Number of unmapped reads = ', unmapped_reads_count)
print('Total number of reads = ', total_number_of_reads)
print('Number of reads with mapping quality 0 = ', zero_mapping_quality_counter)
print('Average mapping quality for all the reads = ', mapping_quality_sum / total_number_of_reads)
print('Average mapping quality without zero quality reads = ', mapping_quality_sum / (total_number_of_reads - zero_mapping_quality_counter))
```
